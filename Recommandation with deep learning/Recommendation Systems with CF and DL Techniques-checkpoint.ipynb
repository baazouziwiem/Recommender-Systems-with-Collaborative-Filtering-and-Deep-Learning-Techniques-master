# Que sont les systèmes de recommandation?

Chaque jour, nous regardons et achetons des produits sur Amazon, naviguons sur Netflix, diffusons de la musique sur Spotify. Vous êtes-vous déjà demandé à quel point ils connaissaient vos goûts et vos préférences et formulaient des recommandations en fonction de vos intérêts?

La réponse est qu’ils ont des moteurs de recommandation personnalisés qui utilisent des algorithmes avancés pour vous recommander des produits ... Lorsque vous souhaitez regarder un nouveau film, vous demandez généralement à vos amis de vous suggérer des films, ils connaissent vos intérêts et vous conseillent par conséquent un bon film. .

Ainsi, **Les systèmes de recommandation fonctionnent de la même manière, ils ne sont que des outils de filtrage de données qui utilisent des algorithmes pour recommander les éléments les plus pertinents à un utilisateur particulier.**

Maintenant, comprenons pourquoi nous avons besoin des systèmes de recommandation et, dans la partie suivante, nous examinerons le fonctionnement réel des systèmes de recommandation.

# Why do we need Recommendation Systems?

Internet est un vaste océan d’informations, c’est des millions d’articles, un très grand catalogue de produits. Certains utilisateurs savent ce qu'ils cherchent, alors que d'autres n'ont aucune idée de ce qu'il faut rechercher dans une aussi grande bibliothèque de ressources. Les systèmes de recommandation jouent donc un rôle essentiel.

Certains produits peuvent être vraiment bons mais n'ont pas gagné **Popularité** car ils n'ont pas été annoncés. Les systèmes de recommandation aident donc ces objets à gagner en popularité en les signalant à leur attention.

Cela peut même aider en **Ciblage des annonces** - Disons que vous cherchez à acheter un nouvel ordinateur portable sur Internet, que vos recherches récentes sont basées sur des suggestions d’ordinateur portable et que bientôt vous verrez des annonces sur des sites Web offrant des réductions sur les ordinateurs portables. Le ciblage des annonces est donc une technique de publicité conçue pour diffuser automatiquement des annonces à l’aide de logiciels spécialisés et d’algorithmes qui les diffusent en fonction des recherches récentes de l’utilisateur. Selon une étude menée en 2009 par la Network Advertising Initiative, le ciblage des annonces aurait généré 2,7 fois plus de revenus que les annonces non ciblées.

Ainsi, cela contribue également à **l'augmentation des revenus** de l'organisation du produit.

# 3) Collaborative Filtering (CF)

**Le filtrage collaboratif** est le processus de filtrage d'informations ou de modèles à l'aide de techniques faisant appel à la collaboration de multiples agents, points de vue et sources de données. Il s'agit essentiellement d'une méthode permettant de faire des prédictions automatiques (filtrage) des intérêts d'un utilisateur en collectant ses préférences. ou goûter les informations de nombreux utilisateurs.

Il existe 2 approches de la FK ->

1) **CF basé sur la mémoire** - Il s'agit d'une approche qui recherche la similarité entre les utilisateurs ou entre les éléments pour recommander des éléments similaires. Les exemples incluent les recommandations relatives aux N principaux basées sur la FC basée sur le voisinage et sur les éléments / utilisateurs.

2) **CF basé sur un modèle** - Dans cette approche, nous utilisons différents algorithmes d'exploration de données et d'apprentissage automatique pour prédire le classement par les utilisateurs d'éléments non notés. Les exemples incluent la décomposition en valeurs singulières (SVD), l'analyse en composantes principales (ACP), etc.

## 3.2) Memory Based Collaborative Filtering

Il existe 2 approches de la CF basée sur la mémoire ->

1) **Filtrage collaboratif user-user** - Dans ce calcul, nous calculons la similarité de tous les utilisateurs avec l'utilisateur actif (l'utilisateur pour lequel la prédiction est destinée) .Ensuite, triez et filtrez les utilisateurs du Top-N afin de utilisateur actif. Ceci est généralement très efficace mais prend beaucoup de temps et de ressources. Par exemple, si Dennis et Davis aiment les mêmes films et qu'un nouveau film plait à Davis, nous pouvons le recommander à Dennis, car Davis et Dennis semblent aimer les mêmes films.

2) **Filtrage collaboratif Item-Item** - C’est semblable à CF user-user, mais nous calculons maintenant la similarité entre les articles pour recommander des articles similaires. Par exemple. Lorsque vous achetez un produit sur Amazon, vous trouverez cette ligne "Les utilisateurs qui ont acheté cet article ont également acheté ...", de sorte qu'Amazon utilise largement l'article CF, bien que je ne dis pas qu'ils n'utilisent que l'article CF, ils ont des techniques hybrides pour mieux répondre aux utilisateurs d’intérêts même uniques.

Item-Item CF sont beaucoup plus rapides que User-User CF. et deuxièmement, les profils des utilisateurs changent rapidement et le modèle de système entier doit être recalculé, alors que les notations moyennes des éléments ne changent pas aussi rapidement, ce qui conduit à des distributions de notation plus stables dans le modèle, de sorte que le modèle ne doit pas être reconstruit. souvent.

**Q) Comment calcule-t-on la similarité? **

**Ans. ** Il existe de nombreuses mesures pour calculer la matrice de similarité, certaines d'entre elles sont ->

1) **Similarité Jaccard** - Il s'agit d'une statistique utilisée pour comparer la similarité et la diversité des ensembles d'échantillons. Elle est définie comme la taille de l'intersection divisée par la taille de l'union des ensembles d'échantillons.

2) **Similarité en cosinus** - Il mesure l'angle entre le vecteur de classement. Si l'angle est égal à 0 °, ce sont des vecteurs de même orientation et si l'angle est de 180 °, il s'agit de vecteurs très dissemblables.

3) **Similarité de Pearson** - Il s'agit en réalité d'une similitude centrée-cosinus. Nous soustrayons les évaluations moyennes des évaluations des utilisateurs, de sorte que la moyenne soit centrée sur 0, puis nous calculons la similarité du cosinus.

### 3.2.1) User based Collaborative Filtering

```python
user_util_matrix = util_mat.copy()

# We will fill the row wise NaN's with the corresponding user's mean ratings, so that we can carry out Pearson correlation.
# Here we assume avg ratings for the movie that is not rated.
user_util_matrix = user_util_matrix.apply(lambda row: row.fillna(row.mean()), axis=1)
user_util_matrix.head(10)
```

```python
user_util_matrix.T.corr()
```

```python
user_corr_mat = user_util_matrix.T.corr()
corr_user_1 = user_corr_mat.iloc[0]
```

Pour des raisons de commodité, je ne considérerai que la corrélation de tous les utilisateurs avec le premier utilisateur uniquement.

```python
corr_user_1.sort_values(ascending=False, inplace=True)
```

```python
corr_user_1
```

```python
# NaN values are generated in corr() as the std dev is zero, which is required in calculating Pearson Similarity.
corr_user_1.dropna(inplace = True)
```

```python
# Neglect the 1st corr value as it is user1 itself
top30_corr_users = corr_user_1[1:51]
```

Ci-dessous, nous avons la liste de tous les produits que l'utilisateur 1 a déjà évalués.

```python
df_combined[ df_combined['userId'] == 1]
```

```python
# user1 has not rated 32 movie
df_combined[ (df_combined['userId'] == 1) & (df_combined['ProductId'] == 32) ] 
```

```python
print('32nd Movie : ', Product['title'][ Product['ProductId'] == 32 ].values)
```

Calculons donc les cotes que l'utilisateur 1 attribuerait au produit à l'aide d'un vecteur similaire. Et sur la base de cette note, nous pouvons la comparer à une note seuil. Si la note est plus élevée, l'utilisateur actif sera visible dans la liste de ses recommandations.

```python
df_n_ratings.loc[['Twelve Monkeys (a.k.a. 12 Monkeys) (1995)']]
```

```python
top30_users = top30_corr_users.keys()

count = 0
users = list()
for user in top30_users:
    if df_combined[ (df_combined['userId'] == user) & (df_combined['ProductId'] == 32) ]['rating'].sum()  :
        count +=1
        users.append(user)

print(count)
```


Il y a 12 utilisateurs similaires parmi les 30 utilisateurs similaires qui ont noté le produit "The Twelve Monkeys".

* Maintenant, calculons la note que l'utilisateur 1 donnerait au produit,


* **Predicted rating** = sum of [ (weights) * (ratings) ]  **/** sum of  (weights)

Here, *weights* est la corrélation de l'utilisateur correspondant avec le premier utilisateur.



```python
# Let's define a function to calculate what user1 will rate the movie
# We use Weighted average of k similar users

def predict_rating():
    sum_similarity = 0
    weighted_ratings = 0
    for user in users:
        weighted_ratings += top50_corr_users.ix[user] * df_combined[ (df_combined['userId'] == user) & 
                                                              (df_combined['ProductId'] == 32) ]['rating'].sum()
        sum_similarity += top50_corr_users.ix[user]

    print(weighted_ratings / sum_similarity)
    
    
predict_rating()
```

```python
df_m[ df_m['ProductId'] == 32]
```

Il y a de très bons changements dans la recommandation de ce produit au premier utilisateur, car la note est plutôt bonne.

Voilà donc comment fonctionne un CF naïf basé sur l'utilisateur. Les notes prévues sont calculées de la même manière pour chaque utilisateur (évidemment pour les films qu'il n'a pas classés) et, en fonction du seuil défini, le film est affiché sur la liste recommandée ou est ignoré.

<a id="there_you_go_4.2"></a>
## 4.2) Building the Model using Embedding Layers

An embedding is a mapping from discrete objects, such as words or movies in our case, to a vector of continuous values. These are used to find similarities between discrete objects. 

The concept behind matrix factorization models is that the preferences of a user can be determined by a small number of hidden factors. And these are called as **Embeddings**.

![Imgur](https://i.imgur.com/zGQJFLD.png)

As you can see in the image, there are 2 features both for user and the items. These are the latent factors or the hidden factors. These factors have a value for the corresponding user and determines to what extend that user likes the feature.

For Eg. the features for the user could be ->
* How much he likes action movies?
* Whether he likes old movies?

And the features for the movies could be ->
* To what scale is it an action movie?
* Whether the movie is recently released?

And finally we take **dot product** which gives us the user's rating for the movie, 

**Estimated Rating** = [ (How much he likes action movies?) x (To what scale is it an action movie) ]  +  [ (Whether he likes old movies?) x (Whether the movie is recently released) ]

```python
# The number of latent factors for the embedding
n_latent_factors = 50

# no of users and movies
n_users, n_movies = len(ratings['userId'].unique()), len(ratings['movieId'].unique()) 
```

<a id="there_you_go_4.3"></a>
## 4.3) Architecture 1

```python
# Model Architecture


# User Embeddings
user_input = Input(shape=(1,), name='User_Input')
user_embeddings = Embedding(input_dim = n_users, output_dim=n_latent_factors, input_length=1, 
                              name='User_Embedding') (user_input)
user_vector = Flatten(name='User_Vector') (user_embeddings)


# Movie Embeddings
movie_input = Input(shape=(1,), name='Movie_Input')
movie_embeddings = Embedding(input_dim = n_movies, output_dim=n_latent_factors, input_length=1, 
                               name='Movie_Embedding') (movie_input)
movie_vector = Flatten(name='Movie_Vector') (movie_embeddings)


# Dot Product
merged_vectors = dot([user_vector, movie_vector], name='Dot_Product', axes=1)
model = Model([user_input, movie_input], merged_vectors)
```

## 4.9) Architecture 2
In this architecture we will be concatenating the latent factors and use Dense layers to calculate the predicted ratings. 

Have a look at the image below ->
![Imgur](https://i.imgur.com/GIe3gTZ.png)

```python
# Model Architecture


# User Embeddings
user_input = Input(shape=(1,), name='User_Input')
user_embeddings = Embedding(input_dim = n_users, output_dim=n_latent_factors, input_length=1, 
                              name='User_Embedding') (user_input)
user_vector = Flatten(name='User_Vector') (user_embeddings)



# Movie Embeddings
movie_input = Input(shape=(1,), name='Movie_Input')
movie_embeddings = Embedding(input_dim = n_movies, output_dim=n_latent_factors, input_length=1, 
                               name='Movie_Embedding') (movie_input)
movie_vector = Flatten(name='Movie_Vector') (movie_embeddings)




# Concatenate Product
merged_vectors = concatenate([user_vector, movie_vector], name='Concantenate')
dense_layer_1 = Dense(100, activation='relu')(merged_vectors) 
# dense_layer_1 = Dropout(0.25) (dense_layer_1)
# batchnorm_layer_1 = BatchNormalization()(dense_layer_1)
# dense_layer_2 = Dense(64, activation='relu')(merged_vectors)


result = Dense(1)(dense_layer_1)
model = Model([user_input, movie_input], result)
```
